### boxplot

```r
# boxplot basico
boxplot(trees$Volume, main='Árvores', xlab='Volume')
```

```r
# personalizando boxplot
boxplot(trees$Volume,
        main = 'Árvores',
        xlab = 'Volume',
        col = 'blue',
        horizontal = TRUE)
```

```r
# plot sem outliers
boxplot(trees$Volume,
        main = 'Árvores',
        xlab = 'Volume',
        col = 'blue',
        outline = FALSE)
```

```r
# intervalo de confiança
boxplot(trees$Height,
        main = 'Árvores',
        xlab = 'Altura',
        col = 'blue',
        notch = TRUE)
```

```r
# estatisticas do boxplot
boxplot.stats(trees$Height)
```

```r
# filtrando estatisticas com boxplot
boxplot.stats(trees$Height)$stats
```

```r
# boxplot para multiplas variaveis
boxplot(trees, main = 'Distribuição de Variáveis das Árvores')
```

### Barras

```r
# preparando dados
spray <- aggregate(. ~spray, data = InsectSprays, sum)
spray
```

```r
# grafico de barras
barplot(spray$count,
        col = gray.colors(6),
        xlab = 'Spray',
        ylab = 'Total',
        names.arg = spray$spray)
box()
```

### Setores

```r
# grafico de pizza
pie(spray$count,
    labels = spray$spray,
    main = 'Distribuição de Spray',
    col = c(1:6))
```

```r
# adicionando legendas ao grafico de pizza
pie(spray$count,
    labels=NA,
    main = 'Distribuição de Sprays',
    col = c(1:6))

legend ("bottomright", legend=spray$spray, fill=c(1:6))
```
