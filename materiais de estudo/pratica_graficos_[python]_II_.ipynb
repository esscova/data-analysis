# Visualizando gráficos com Python - parte II
## Gráficos de Dispersão

```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# carregando dados
df_arvores = pd.read_csv("../datasets/trees.csv", sep=',', encoding='utf-8')
df_co2 = pd.read_csv('../datasets/co2.csv', sep=',', encoding='utf-8')
```

```python
df_arvores.head()
```

```python
df_co2.head()
```

```python
# pandas
df_arvores.plot(
    kind = 'scatter',
    x = 'Girth',
    y = 'Height',
    color = 'blue',
    s = 80)
plt.title('Relação entre Diâmetro e Altura das Árvores')
plt.xlabel('Diâmetro da Árvore (cm)')
plt.ylabel('Altura da Árvore (cm)')
plt.show()
```

```python
# matplotlib
plt.scatter(df_arvores['Girth'], df_arvores['Height'], s=100, c='green')
plt.title('Relação entre Diâmetro e Altura das Árvores')
plt.xlabel('Diâmetro da Árvore (cm)')
plt.ylabel('Altura da Árvore (cm)')
plt.show()
```

```python
# seaborn
sns.scatterplot(x='conc', y='uptake', data=df_co2)
plt.title('Relação entre Concentração de CO2 e Taxa de Absorção')
plt.xlabel('Concentração de CO2')
plt.ylabel('Taxa de Absorção')
plt.show()
```

```python
# personalizando seaborn
sns.scatterplot(x='conc', y='uptake', hue='Treatment',data=df_co2)
plt.title('Relação entre Concentração de CO2 e Taxa de Absorção')
plt.xlabel('Concentração de CO2')
plt.ylabel('Taxa de Absorção')
plt.show()
```

```python
# dispersão com linha de regressao
m,b = np.polyfit(df_arvores['Girth'], df_arvores['Height'], 1)
plt.scatter(df_arvores['Girth'], df_arvores['Height'], color='blue', s=80)
plt.plot(df_arvores['Girth'], m*df_arvores['Girth'] + b, color='red')
plt.title('Relação entre Diâmetro e Altura das Árvores com Linha de Regressão')
plt.xlabel('Diâmetro da Árvore (cm)')
plt.ylabel('Altura da Árvore (cm)')
plt.show()
```
