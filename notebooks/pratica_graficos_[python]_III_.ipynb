# Visualizando dados com Python - Parte III
## Subplots

```python
# bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gerando dados
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 4]
```

```python
# 1 linha x 2 colunas
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# plot do seno
axes[0].plot(x, y1, color='blue')
axes[0].set_title('Gráfico de Seno')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# barplot
axes[1].bar(categories, values, color='green')
axes[1].set_title('Gráfico de Barras')
axes[1].set_xlabel('Categoria')
axes[1].set_ylabel('Valor')

plt.tight_layout()
plt.show()
```

```python
# 1 linha x 3 colunas
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# barplot
axes[0].bar(categories, values, color='green')
axes[0].set_title('Gráfico de Barras')
axes[0].set_xlabel('Categoria')
axes[0].set_ylabel('Valor')

# seno
axes[1].plot(x, y1, color='blue')
axes[1].set_title('Gráfico de Linhas (Seno)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

# scatter cosseno
axes[2].scatter(x, y2, color='red')
axes[2].set_title('Gráfico de Dispersão (Cosseno)')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')

plt.tight_layout()
plt.show()
```

```python
# 2 linhas x 2 colunas
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# seno
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title('Gráfico de Seno')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')

# cosseno
axes[0, 1].plot(x, y2, color='red')
axes[0, 1].set_title('Gráfico de Cosseno')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')

# tang
axes[1, 0].plot(x, y3, color='green')
axes[1, 0].set_title('Gráfico de Tangente')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].set_ylim(-10, 10) # evita valores infinitos

# scatter
axes[1, 1].scatter(x, y2, color='purple')
axes[1, 1].set_title('Gráfico de Dispersão')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')

plt.tight_layout()
plt.show()
```
