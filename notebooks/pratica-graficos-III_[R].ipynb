# Praticando graficos em R com Lattice

```r
# carregando biblioteca lattice
library(lattice)
```

```r
# boxplots
bwplot(trees$Volume,
       main = 'Distribuição do Volume de Árvores',
       xlab = 'Volume')    
```

```r
# histogramas
histogram(trees$Volume,
          main = 'Distribuição do Volume de Árvores',
          xlab = 'volume',
          aspect = 1,
          type = 'percent',
          nint = 5)
```

```r
# histogramas condicionais
histogram(~weight | feed,
          data = chickwts,
          main = 'Peso Condicionado à Alimentação',
          xlab = 'Peso')
```

```r
# dispersao com condicionais
xyplot(CO2$conc ~ CO2$uptake | CO2$Type,
       main = "Concentração vs Captação de CO2 por Tipo",
       xlab = 'Captação de CO2',
       ylab = 'Concentração de CO2')
```

```r
xyplot(CO2$conc ~CO2$uptake | CO2$Treatment,
       main = 'Concentração vs Captação de CO2 por Tratamento')
```

```r
# matrizes de dispersão
splom(~CO2[4:5] | CO2$Type, main = 'Matriz de Dispersão por Tipo de Planta')
```

```r
# densidade
densityplot(~CO2$conc | CO2$Treatment, 
            plot.points = FALSE,
            main = 'Densidade da Concentração de CO2 por Tratamento')
```

```r
# 3d
cloud(decrease ~rowpos * colpos,
      groups = treatment,
      data = OrchardSprays,
      main = 'Impacto de Sprays em Abelhas')
```

```r
# level plots
levelplot(Girth ~ Height * Volume,
          data = trees,
          main = 'Relação entre Circunferência, Altura e Volume')
```
