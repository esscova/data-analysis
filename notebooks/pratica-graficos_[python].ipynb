# Visualizando gr√°ficos com Python

```python
# bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# carregando dados
df = pd.read_csv('../datasets/trees.csv', sep=',', encoding='utf-8')
df.head()
```

```python
# info
df.info()
```

## Histogramas

```python
# obtendo dados do intervalo histograma com numpy
np.histogram(df.iloc[:,1], bins=6)
```

```python
# plot com pandas
df.iloc[:,1].hist(bins=6)
```

```python
# Histograma 
df['Height'].plot(kind='hist', bins=10, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribui√ß√£o das Alturas das √Årvores')
plt.xlabel('Altura (metros)')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# plot com matplotlib
plt.hist(df['Height'], bins=6)
plt.title('Distribui√ß√£o de Frequ√™ncia')
plt.xlabel('Altura')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# plot com seaborn
sns.histplot(data=df['Height'], bins=6).set_title('Distribui√ß√£o de Frequ√™ncia')
plt.xlabel('Altura')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# seaborn com kde
sns.histplot(df['Height'], kde=True)
plt.title('Distribui√ß√£o das Alturas das √Årvores com KDE')
plt.xlabel('Altura (metros)')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# kde com suaviza√ß√£o
sns.histplot(df['Height'], kde=True, kde_kws={'bw_adjust':0.5})
plt.title('Distribui√ß√£o das Alturas das √Årvores com KDE (bw_adjust=0.5)')
plt.xlabel('Altura (metros)')
plt.ylabel('Frequ√™ncia')
plt.show()
```
